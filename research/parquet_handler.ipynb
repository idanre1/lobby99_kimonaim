```python
import pandas as pd
import numpy as np
import sqlite3
from tqdm.notebook import tqdm
```

```python
# establish database connection
conn = sqlite3.connect('/datadrive/lobby99/download/shufersalist.db')
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type ='table' AND name NOT LIKE 'sqlite_%'", conn)
tables=tables['name'].tolist()
tables
```

```python
priceTables=['shufersalPrices',
 'ramiLeviPrices',
 'victoryPrices',
 'osherAdPrices',
#  'yohananofPrices',
 'yenotBitanPrices',
 'yohananof1Prices']
df=pd.DataFrame()
with tqdm(total=len(priceTables)) as pbar:
    for p in priceTables:
        pbar.set_description('Processing: %s' % p)
        pbar.update(1)
        q=pd.read_sql_query("SELECT * FROM %s" % p, conn)
        q['vendor']=p
        df=df.append(q)
df
```

```python
df.reset_index(drop=True, inplace=True)
df.to_parquet('/datadrive/lobby99/kimonaim.parq')
```
